```python
df_profit.reset_index(inplace=True)
df_profit = df_profit.drop(columns=df_profit.columns[0], axis=1)
```

```python
df_profit['gross_profit_usd'].fillna(0, inplace=True)
```

```python
gross_profit = [0]*4
miner_payment = [0]*4
mev = [0]*4

for i in range(len(df_profit)):
    if 12965000 <= df_profit['block_number'][i] < 12973750:
        gross_profit[0] += df_profit['gross_profit_usd'][i]
        miner_payment[0] += df_profit['miner_payment_usd'][i]
        mev[0] += (df_profit['gross_profit_usd'][i] - df_profit['miner_payment_usd'][i])
        
    elif 12973750 <= df_profit['block_number'][i] < 12982500:
        gross_profit[1] += df_profit['gross_profit_usd'][i]
        miner_payment[1] += df_profit['miner_payment_usd'][i]
        mev[1] += (df_profit['gross_profit_usd'][i] - df_profit['miner_payment_usd'][i])
        
    elif 12982500 <= df_profit['block_number'][i] < 12991250:
        gross_profit[2] += df_profit['gross_profit_usd'][i]
        miner_payment[2] += df_profit['miner_payment_usd'][i]
        mev[2] += (df_profit['gross_profit_usd'][i] - df_profit['miner_payment_usd'][i])
        
    elif 12991250 <= df_profit['block_number'][i] < 13000000:
        gross_profit[3] += df_profit['gross_profit_usd'][i]
        miner_payment[3] += df_profit['miner_payment_usd'][i]
        mev[3] += (df_profit['gross_profit_usd'][i] - df_profit['miner_payment_usd'][i])
```
